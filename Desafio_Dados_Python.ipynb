```python
#media idade
mediaidade = df['Age'].mean()
print("A média das idades dos funcionários é:", mediaidade)
```

```python
#media de tempo na empresa
mediatempo = df['YearsAtCompany'].mean()
print("A média de tempo dos funcionários na empresa é:", mediatempo)
```

```python
#media do nível de satisfação no trabalho:
mediasatisfacao = df['JobSatisfaction'].mean()
print("A média do nível de satisfação no trabalho é:", mediasatisfacao)
```

```python
df2 = df.groupby('EducationField')[['DailyRate']].mean()
df2.plot(kind = 'bar')
```

```python
df2 = df.groupby('Department')[['YearsAtCompany']].mean()
df2.plot(kind = 'bar')
```

```python
department_counts = df['Department'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Distribuição de Funcionários por Departamento')
plt.axis('equal')
plt.show()
```

```python
df_line = df.groupby('TotalWorkingYears')[['DailyRate']].mean()
plt.figure(figsize=(10, 6))
plt.plot(df_line, marker='o', linestyle='-', color='b')
plt.title('Salário Diário Médio por Total de Anos Trabalhados')
plt.xlabel('Total de Anos Trabalhados')
plt.ylabel('Média do DailyRate')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Age', 'YearsAtCompany', 'TotalWorkingYears', 'DailyRate']], diag_kind='kde')
plt.show()
```
